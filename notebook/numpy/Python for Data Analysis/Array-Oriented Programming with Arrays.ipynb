```python
import numpy as np
```

```python
import matplotlib as plt
```

```python
points = np.arange(-5, 5, 0.01)
```

```python
xs, ys = np.meshgrid(points, points)
```

```python
ys
```

```python
z=np.sqrt(xs**2+ys**2)
```

```python
z
```

```python
import matplotlib.pyplot as pyplot
```

```python
pyplot.imshow(z, cmap=plt.cm.gray); pyplot.colorbar(); pyplot.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```
